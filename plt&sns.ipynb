##### Numerical data == Quantative data

```python
# We use distribution plots for quantativa dataset
# 1. distplot/histplot
# 2. jointplot
# 3. pairplot
# 4. rugpot
```

```python
#  So first import libraries as always:-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# So seaborn have their own data, which i used for practice:-
# if you want to use any other data , you can upload it by using pandas
# so lets start with the sample data of seaborn:-
#  and simply i load dataset like this:-
df = sns.load_dataset('tips')
df
```

```python
df.shape
```

```python
# so first I checked the size columns , how many uniques values in this column
df['size'].unique()
#  so see there are only 6 unique values in that column, 
#  so i can consider this column as qualitative data
```

```python
#  so only total bills and tips is only two quantative column
# so we use this two column for creating distribution plot
sns.displot(df['total_bill'])
```

```python
# we can also pass bins value in parameter , means how many no of bars you need in your plot
# example
sns.displot(df['total_bill'],bins = 5)
```

```python
# We also pass kde parameter 
# kernel density estimator which show the variance in our plot, show skewness in the plot (Left skewed or right skewed)
# example-
sns.displot(df['total_bill'],bins= 20,kde=True)
```

```python
# so can we make plot for both column - total bills and tips
#  yes we can use matpltlib here
# see how
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins= 20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins= 20,kde=True)

# There is no difference in histplot and distplot
```

```python
# JOINT PLOT:-
# we can pass different column on the axis 
# like xaxis= total_bill and yaxis= tip
# In the above plots , we won't be able to comaprison between two columns 
# What if I want to compare two column, so here we use joint plot:-
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
# Here i pass total bill column in x axis and tip column in y axis, dataframe is 'df' and type of plot in kind(by default is scatter)
```

```python
# whatif i choose any other type of plot in kind, see
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
# PAIR PLOT:-
# We can directly pass our data frame 
sns.pairplot(df)
#  Its give a pair of plot of all quantative column
```

```python
sns.pairplot(df,hue='sex')
# here i pass hue= sex, which is qualitative column and its divide the tip category wise (male/female)
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
# now see that on the basis of size and we also pass palette for designing
```

```python
# RUGPLOT:-
# we can directly pass whole dataframe or any column
sns.rugplot(df['total_bill'])
# here only we can see that , how many values lie
```

```python
sns.rugplot(df)
#  pass the whole dataframe, how many values of total bill, tip and size lie in our range.
```

##### Categorical Data == Qualitative data

```python
#  Now let's see , use of seaborn to plot categorical data! 
# There are few main plots :
# 1. boxplot
# 2. violinplot
# 3. stripplot
# 4. swarmplot
# 5. barplot
# 6. countplot
```

```python

df1 = sns.load_dataset('tips')
df1
#  so there are som columns of categorical data, like
#  sex, smoker, day and  time
```

```python
# COUNTPLOTS:---

# So first we discuss about count plots:-
# why , because we don't know , how many no. of males and females are in the sex column or no.of yes/no in smoker column and so on
#  so for that
sns.countplot(df['sex'])
# here i simply pass my dataframe of only sex column , becasue i only want to get count of that colomn first.
# Here it makes a plot on y axis , because its by default
```

```python
# if i want that plot on x axis , i simply pass that column in x 
sns.countplot(x=df['sex'])
```

```python
# Now I want to divide that genders on the basis of smoking
#  so simply I pass datframe with smoker column in hue like this
sns.countplot(x=df['sex'],hue=df['smoker'])

#  So its show us , there are apprx 100 males who are non smoker and around 60 males are smoker, same as for females.
```

```python
# BARPLOTS:--------------
# So the above plots also looks like barplot,but
# In the barplots , it takes two positional arguments
# Need two column values for X axis and Y axis respectively
# But on Y axis , we have to pass numerical data like tip, total bills or size
# And on the x axis , we have to pass categorical data like sex, smoker, day and so on

sns.barplot(x= df['sex'], y=df['total_bill'] )
#  here its show me the average of total bill which is divided in between sex's 
#  By default it take avaerage estimator
```

```python
# We can also pass estimator by our choice:- see
sns.barplot(x= df['sex'], y=df['total_bill'], estimator=np.sum )
# see i pass sum of total bills which is pay by males and females
```

```python
# BOXPLOTS:------------ IMPORTANT
# Boxplot create on the basis of 5 point summary, mainly used for finding outliers

sns.boxplot(x='tip',y= 'day',data= df1, palette= 'rainbow')
```

Boxplots and violinplots are used to shown the distribution of categorical data. A box plot shown the distribution of quantative data in a way that facilitates comparisons between variables or across level of a categorical variable. The box shown the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are destermined to be 'outliers' using a method that is a function of inter-quartile range.

```python
# VIOLINPLOTS:-----------

sns.violinplot(x='day',y='total_bill', data= df1)
# So here its show the variance of data , 
#  for example:- see on the day 'thursday' , its spread in the range of 10-20 oof total bills , similar for others
```

```python
# STRIPPLOT:-
# its drawn a scatterplot where one variable is categorical.
sns.stripplot(x='day',y='total_bill',data= df1)

# So you can see , there is scatter strip pn the day basis 
```

```python
# SWARMPLOTS:-----
# Its similar to stripplot , but the points are adjusted , so that they don;t overlap.
# Its use for better distribution of values 

sns.swarmplot(x='day',y='total_bill',data= df1)
#  you can see , its show me the same plot like stripplot , but the datapoints are spreaded in this plot, not overlap each other,
# its also similar to violin plot 
```

```python
# So can we join swarmplot and violinplot :-
# yes, how??   see,,,
sns.violinplot(x='tip',y='day',data=df1)
sns.swarmplot(x='tip',y='day',data=df1)
```

```python
#  HEAT MAP:-------
#  So first discuss heat map
# what it this?
# Heat map is creating on the basis of correlation.
# What is correlation
# so lets consider , we have three columns
#    A          B          C
#    1        1000        100
#    2         900        200
#    3         500        300 
#    4         400        450
#    5         300        550

# Now first thing, corralation always lie in between -1 to 1
#  If I want to predict the value of c , so I have to know baout A and B
# If there is any relation between A and C, we can say that is correaltion , similar for B also. 
# You can see If the value of A increased , the value of c is also increased , this is called +ve correaltion in bw A and c
#  On the other hand , ifthe value of B is decreased, the value of C is increased, this is called  -ve correlation.
# How we can find that correation
# diving  covariance by standard deviation is known as correlation
```

```python
#  So lets see how we can calculate correlatiom:
#  so here i take tip data , there is not any role of numerical or categorical
#  you always need a data of integer type.
#  So in colum total_bill, tip and size have integer data
#  so i can take that columns
tipscorr = tips[['total_bill','tip','size']]
tipscorr
```

```python
# Now i want to find correlation in b/w total bill and size, tips and size , and so on.
tipscorr.corr()
#  So here its provide me relation between all of them
```

```python
#  Now see how can we use this for creating heatmap:-
# simply pass the above correlation in heatmap and see the plot
sns.heatmap(tipscorr.corr())

#  so as you see , its show me heatmap in matrix form , that why its also known as matrix plot
# we have all +ve correation, that's why its showing me like this
#  if i have -ve correlation, it s show that too.
```

```python
#  I also pass the value inside that boxes , by simply pass annotation 
# like this:-
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
# CLUSTER MAP:-------
#  there is not so much difference in heatmap and cluster map
# Its same as heat map , but its show distribution tree and continous distribution only.
# see 
sns.clustermap(tipscorr.corr())
```

```python
# PIVOT TABLE : HEATMAP
# Now see how we can make heatmap by using pivot table
# so here i use my flights data:
# let's see
pvtflights = flights.pivot_table(values='passengers',index='month',columns='year')
pvtflights

#  so see by using pivot table, i distribute my data as i needed
```

```python
#  now see , how can i create a heatmap by using that pivot table data
#  why i do this, for finding , in which year and month has highest and lowest no of flights and 
sns.heatmap(pvtflights)

# so see  , there are high no of flights in 1960 in the month of aug and july, lowest no. of flights in 1949 in jan and dec.

```

```python
# Regression plot used ML algorithm, not used so much , but atleast we will see how its works
# So as we know ,we show our prediction on y axis and on the x axis put that required data which is used for prediction
# Simple:- Consider you have any scatter plot and one linear line passed over there , this is called regression plot
#  for creating this plot , we use lmplot()
#  see , how.....,
# I want to predict how much tip will be given on the basis of total bill
tips = sns.load_dataset('tips')
sns.lmplot(x = 'total_bill',y = 'tip', data = tips)

```

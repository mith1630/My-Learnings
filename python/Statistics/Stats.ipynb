## Statistics for DS:-

```python
# Code Implementation :-
# Measure of central Tendency:- mean, median and mode
# Measure of spread :- Quartiles and IQR
# Outliers :- Upper Fence (Q3 + 1.5(IQR)) AND Lower Fence (Q1 - 1.5(IQR))
# Five Number summary :- Min , Q1 , Q2 , Q3 , Max (Taking this five number after removing outliers)

import numpy as np
import seaborn as sns
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
# So here we want to remove outlier and for that I have to find quartiles first
# How can we do this: see......,
# We already have a function percentile in numpy libraries
# so simply i pass my array in that function with quartile percentage (Q1-25, Q2-50, Q3-75)
# I will pass it, like this:-

Q1 = np.percentile(arr,25)
Q2 = np.percentile(arr,50)
Q3 = np.percentile(arr,75)
Q3
```

```python
# Now lets find the IQR : (IQR = Q3 - Q1)
IQR = Q3 - Q1
IQR
```

```python
# After that , we need upper fence and lower fence
UF = Q3 + (1.5*IQR)
UF
```

```python
LF = Q1 - (1.5*IQR)
LF
```

```python
# So in our dataset , the numbers which is lie in b/w LF to UF is needed, apart from this range all the numbers count as OUTLIERS.
# And we remove that outliers from our dataset:-
# so we use for loop here and create a new list to store the new dataset

l = []
for i in arr:
    if i >= LF and i <= UF:
        l.append(i)

l       
# So here i get the new list of data without outlier
# as we know 201 is the outlier, and its removed 
```

```python
# now i want that list in array format
# so simply i pass that list in our new dataset variable arr2
arr2 = np.array(l)
arr2
```

```python
# What if I create box plot of my both arrays (arr and arr2):-
# let's see:-
sns.boxplot(x=arr)
# See when i create a box plot with outliers , its doesn't look like in proper way
```

```python
# But when i create a boxplot without outlier:-
# see how its look:-
sns.boxplot(x=arr2)

# And this boxplot properly show our five no. summary


# Even you can see that in our plot:-
# 1st vertical line from LF-5 is our minimun 
# 2nd vertical line from 5-10 is our Q1
# 3rd vertical line from 10-12 is our Q2 whichis also called mean
# 4th vertical line from 20-25 is our Q3
# And the last vertical line from 35-UF is our maximum
```

```python
# Now lets move to the questions part:-
# 1. Which country has the highest population?
# How can I do that :- By conditional selection...
# df['population'] == df['population'].max()
# So first we have to target the dataframe of population and make condition of maximum
# After doing that, put that condition in whole database.
# How??
# see/
df[df['population'] == df['population'].max()]
```

```python
# So India has the highest population:-
# If you only want name of the country, simply add column name at the end of ops
df[df['population'] == df['population'].max()]['country']
# So here I got the name of the country only.
```

```python
# 2. What is the capital of the country with highest population?
# Simply add column name of capital city in the previous command instead of country
df[df['population'] == df['population'].max()]['capital_city']
# That's it, so New Delhi is the only capital who have highest population.
```

```python
# 3. Which country has the least population?
# Again I use the previous command, but here I use min() function instead of max(). 
df[df['population'] == df['population'].min()]['country']
# So here Tuvalu has the least popultion
```

```python
# 4. What is the capital of the country with least population?
# Again instaed of max , i use min() function.
df[df['population'] == df['population'].min()]['capital_city']
# So Funafuti is the  only capital who have least population.
```

```python
# 7. How many countrues lie in Eastern Europe region.
df[df['region'] == "Eastern Europe"]['country']
# Simply I take my dataset df and region column == Eastern Europe, and add country column name at the end to know the countries name.
```

```python
# 8. Who is the political leader of the 2nd highest populated country?
# df[df['population'] == df['population'].nlargest(2)]
# if i do like the above command  its giving me a error
# Can only compare identically-labeled Series objects
# So what can i do?
# I use iloc() here and give a index value 1.
# Because nlargest function give multiple answer for this,
# And we know that 1st country is India who have highest population
# so logically india lie on the 0th index and we want 2nd largest, so we pass 1th index in iloc funcyion.
# df[df['population'] == df['population'].nlargest(2).iloc[1]]
# So its show me china on the 2nd position, but i want the name of political leader
# So simply I add political leader column name at the end.
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
#  Xi Jinping is the political leader with 2nd highest populated country.
```

```python
# 9. How many countries are there whoes political leader are unknown?
df[df['political_leader'].isna()]['country']
# So there 7 countries whoes political leader are unknown.
```

```python
# 10. How many countries have Republic in their full name?
# IMPORTANT QUESTION ; CODE NEEDED
# Here we need to create function
count = 0          # Counting start from zero
def counting(txt):
    global count   # make count variable global, because i have to use it in my function
    if 'republic' in txt.lower():  # get the txt in lowercase each time
        count +=1 # if word 'republic' exist , increment of 1 in count 
    return txt

df['country_long'] = df['country_long'].apply(counting) # simply pass my function in apply
print(count) # Print the total no. of counts

# So there are 125 countries are republic.
```

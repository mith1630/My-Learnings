#### MISSING DATA:-

```python
import numpy as np
import pandas as pd
```

```python
# FIND MISSING DATA
data = {
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Now i use isna fumction to know null values in boolean
df.isna()
```

```python
# Now I want to count all nullvalues:-
df.isna().sum()
```

```python
#  now i want to , all the rows have null value or any of the rows have all values
# so i use any() here to know that:-
df.isna().any()

# so here all the rows have nullvalues
```
